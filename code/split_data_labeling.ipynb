```python
import os
import pandas as pd
import numpy as np
import uuid
```

```python
_DATA_PATH = '../data'
_RESOURCE_PATH = '../resource'
```

```python
# df_data_labeling = pd.read_csv(os.path.join(_DATA_PATH,'data-QA.csv'))
df_data_labeling = pd.read_csv(os.path.join(_DATA_PATH,'title.csv'))
df_data_labeling
```

```python
# del df_data_labeling['Unnamed: 0']
```

```python
# df_data_labeling = df_data_labeling.rename(columns={'question':'Câu hỏi'})
df_data_labeling = df_data_labeling.rename(columns={'title':'Câu hỏi'})
```

```python
df_data_labeling[['Câu hỏi']].head(1)
```

```python
len(list(set(list(df_data_labeling['Câu hỏi']))))
```

```python
df_data_labeling = df_data_labeling.drop_duplicates()
len(df_data_labeling)
```

```python
start_point = 0
num_record = 11
# space = 1600
space = 900
tasks = ['row_based_may19','token_based_may19']
len_id = len(str(len(df_data_labeling)))

list_des = []

for task in tasks:
    folder_task = os.path.join(_RESOURCE_PATH,task)
    if not os.path.isdir(folder_task):
        os.mkdir(folder_task)
        
    if task == 'row_based_may19':
        extension = '.csv'
    else:
        extension = '.txt'
        
    for idx in range(num_record):
        
        dict_des = {}
        
        if idx < num_record - 1:
            
            start = idx*space
            end = (idx+1)*space
            
            curr_df = df_data_labeling[start:end]
            

            start_str = str(start)
            if len(start_str) < len_id:
                start_str = '0'*(len_id-len(start_str)) + start_str
            
            end_str = str(end)
            if len(end_str) < len_id:
                end_str = '0'*(len_id-len(end_str)) + end_str
            
            task_id = task[0]+start_str[:-2]+end_str[:-2]
            
            trace = str(idx*space)+'_'+str((idx+1)*space)
            
            dict_des['task'] = task
            dict_des['_id'] = task_id
            dict_des['index'] = trace
            
            list_des.append(dict_des)

            if task == 'row_based_may19':
                curr_df[['Câu hỏi']].to_csv(os.path.join(folder_task,task_id + extension),header=True,index=False)
            else:
                curr_df[['Câu hỏi']].to_csv(os.path.join(folder_task,task_id + extension),header=False,index=False)
        else:
            curr_df = df_data_labeling[idx*space:]     
            
            start = idx*space
            end = len(df_data_labeling)
            
            curr_df = df_data_labeling[start:end]
            

            start_str = str(start)
            
            end_str = str(end)

            task_id = task[0]+start_str[:-2]+end_str[:-2]
            
            trace = str(idx*space)+'_'+str(len(df_data_labeling))
            
            dict_des['task'] = task
            dict_des['_id'] = task_id
            dict_des['index'] = trace
            
            list_des.append(dict_des)
            if task == 'row_based':
                curr_df[['Câu hỏi']].to_csv(os.path.join(folder_task,task_id + extension),header=True,index=False)
            else:
                curr_df[['Câu hỏi']].to_csv(os.path.join(folder_task,task_id + extension),header=False,index=False)
```

```python
df_des = pd.DataFrame(list_des)
df_des.to_csv(os.path.join(_RESOURCE_PATH,'project_labeling_tracking_may19.csv'),header=True,index=False)
```
