```python
list_cols = list(dataset.columns)
list_cols.pop(0)
list_merge_full = []
for idx in range(len(dataset)):
    if idx < len(dataset):
        df = dataset[idx:idx+1]
    else:
        df = dataset[idx:]
        
    list_merge_row = []
    for col in list_cols:
        symp = list(df[col])[0]
        if symp not in list_merge_row:
            list_merge_row.append(symp)
    list_merge_full.append(list_merge_row)
```

```python
dataset['merge_symp'] = list_merge_full
```

```python
# list_cols.pop(0)
```

```python
dataset.head()
```

# GROUPBY

```python
dataset_groupby = dataset.groupby('Disease')['merge_symp'].apply(list).reset_index(name='groupby_symp')
```

```python
dataset_groupby.head()
```

```python
def flatten_list_of_list(lol):
    sublist = []
    flat_list = [item for sublist in lol for item in sublist]
    return flat_list
```

```python
list_groupby_symp_flatten = []

for item in list(dataset_groupby['groupby_symp']):
    flatten_symp = flatten_list_of_list(item)
    unique_symp = list(set(flatten_symp))
    list_groupby_symp_flatten.append(unique_symp)
    
dataset_groupby['groupby_symp_flat'] = list_groupby_symp_flatten
```

```python
dataset_groupby.head()
```

# DIFFERENT

```python
list_disease = list(dataset_groupby['Disease'])
# list_disease


list_invest = []

for d1 in list_disease:
#     dict_invest[d1] = {}
    dict_invest = {}
    dict_invest['Disease'] = d1
    
    df1 = dataset_groupby[dataset_groupby['Disease']==d1]
    list_symp_origin = list(df1['groupby_symp_flat'])[0]
    
    list_diff = []
    
    for d2 in list_disease:
#         print(d1,'---',d2)
#         if d1 != d2:
        df2 = dataset_groupby[dataset_groupby['Disease']==d2]

        list_symp_compare = list(df2['groupby_symp_flat'])[0]

        diff = len(set(list_symp_origin) - set(list_symp_compare))

#             dict_invest[d1][d2] = diff
        list_diff.append([diff,d2])
    dict_invest['Different'] = list_diff
    list_invest.append(dict_invest)
```

```python
# list_invest
```

```python
df_invest = pd.DataFrame(list_invest)
# df_invest
```

```python
def visualize_count_symptom(list_count,disease):
    plt.figure(figsize=(10,10))
    
    list_count_sort = sorted(list_count,reverse=False)
    
    plt.barh([item[1] for item in list_count_sort],[item[0] for item in list_count_sort])
    plt.ylabel(disease)
    plt.xlabel('Quantity')
    plt.show()
```

```python
for disease in list_disease:
    list_count = list(df_invest[df_invest['Disease']==disease]['Different'])[0]
    visualize_count_symptom(list_count,disease)
```
