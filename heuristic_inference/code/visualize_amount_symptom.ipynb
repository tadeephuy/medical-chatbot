```python
dataset = dataset.fillna('none_symptom')
```

# GROUPBY

```python
list_symptom_columns = list(dataset.columns)
list_symptom_columns
```

```python
list_symptoms_unique = []
for col in list_symptom_columns:
    list_symptoms = list(dataset[col].unique())
    for item in list_symptoms:
        if item not in list_symptoms_unique:
            list_symptoms_unique.append(item)
```

# COUNT

```python
dict_count = {}
for col in list_symptom_columns:
    if col != 'Disease':
        list_count_one_col = []

        list_symptoms = list(dataset[col])
        for item in list_symptoms_unique:
            
            if list_symptoms.count(item) != 0:
                zip_count_one_symptom = [list_symptoms.count(item),item]
                list_count_one_col.append(zip_count_one_symptom)
        zip_count_sort = sorted(list_count_one_col,reverse=False)
        dict_count[col] = zip_count_sort
```

```python
def visualize_count_symptom(list_count,symptom):
    plt.figure(figsize=(10,10))
    plt.barh([item[1] for item in list_count],[item[0] for item in list_count])
    plt.ylabel(symptom)
    plt.xlabel('Quantity')
    plt.show()
```

```python
for symp in list(dict_count.keys()):
    list_count = dict_count[symp]
    visualize_count_symptom(list_count,symp)
```
